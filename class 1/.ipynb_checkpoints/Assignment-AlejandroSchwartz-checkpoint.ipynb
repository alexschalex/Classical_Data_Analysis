### Simple linear regression

````python
# Prepare data for training

#PUT YOUR CODE HERE

## Add further cells in case you need

X=customers_clean[['Length of Membership']]
Y=customers_clean['Yearly Amount Spent']
````

### Training and test splits
Split the data into training and testing sets. We train the model with 75% of the samples and test with the remaining 25%. We do this to assess the models performance on unseen data. To split the data you may use ```scikit-learn library```. Finally print the sizes of our training and test set to verify if the splitting has occurred properly.

**Use model_selection.train_test_split from sklearn to split the data into training and testing sets. Set random_state=5**

````python
#PUT YOUR CODE HERE

## Add further cells in case you need
from sklearn.model_selection import train_test_split

X_train,X_test, Y_train, Y_test=train_test_split(X,Y,test_size=.25,random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
````

### Training of the model

````python
#PUT YOUR CODE HERE

## Add further cells in case you need
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
````

#### Print the coeficients of the model

````python
#PUT YOUR CODE HERE

## Add further cells in case you need



lin_model.intercept_
````

````python
coeff_df = pd.DataFrame(lin_model.coef_,X.columns,columns=['Coefficient'])
coeff_df
````

### Make a plot to show how the model fit the training data.

Looking at the plot explain if the coefficients you found make sense

````python
#PUT YOUR CODE HERE

## Add further cells in case you need

plt.scatter(X_train, Y_train, color = "blue")
plt.plot(X_train, lin_model.predict(X_train), color = "green")
plt.title("LofMem vs Yearly Spent (Training set)")
plt.xlabel("L of Mem")
plt.ylabel("Yearly Spent")
plt.show()
````

So the intercept is 270, that would mean that a client with no time of membership (a new client) would spent 270Euros on the first Year. After the first, this would increase by 65 Euros every year that they stay being as clients.

### Evaluate the model on the test set

    - Make some plots you think maybe usefull to evaluate the predictions of our model and then evaluate it through some metrics.

#### Evalution Metrics

````python
#PUT YOUR CODE HERE

## Add further cells in case you need
predictions = lin_model.predict(X_test)


````

````python
plt.scatter(Y_test,predictions)
````

````python
sns.distplot((Y_test-predictions),bins=50);
````

Both plots shows that the prediction in not that bad, but there is some noise (the errors goes between -100 and 100 Euros). They are quite good distributed, but the model looks that it could be improved.

````python
from sklearn import metrics

y_train_predict = lin_model.predict(X_train)

print("The model performance for training set")
print('MAE:', metrics.mean_absolute_error(Y_train, y_train_predict))
print('MSE:', metrics.mean_squared_error(Y_train, y_train_predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_train, y_train_predict)))
print('R2 Score:', metrics.r2_score(Y_train, y_train_predict))

print("\n")

y_test_predict = lin_model.predict(X_test)
print("The model performance for testing set")
print('MAE:', metrics.mean_absolute_error(Y_test, y_test_predict))
print('MSE:', metrics.mean_squared_error(Y_test, y_test_predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, y_test_predict)))
print('R2 Score:', metrics.r2_score(Y_test, y_test_predict))
````

Here we can observe that the model is explainig a 65% approx of the data, there space to improve, but is important that the model is staiyn stable between train and test sets.

````python
X = pd.DataFrame(np.c_[customers_clean['Avg. Session Length'], customers_clean['Time on Website'],customers_clean['Length of Membership']], columns = ['Avg. Session Length','Time on Website','Length of Membership'])
Y = customers_clean['Yearly Amount Spent']
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
````

````python
y_train_predict = lin_model.predict(X_train)

print("The model performance for training set")
print('MAE:', metrics.mean_absolute_error(Y_train, y_train_predict))
print('MSE:', metrics.mean_squared_error(Y_train, y_train_predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_train, y_train_predict)))
print('R2 Score:', metrics.r2_score(Y_train, y_train_predict))

print("\n")

y_test_predict = lin_model.predict(X_test)
print("The model performance for testing set")
print('MAE:', metrics.mean_absolute_error(Y_test, y_test_predict))
print('MSE:', metrics.mean_squared_error(Y_test, y_test_predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, y_test_predict)))
print('R2 Score:', metrics.r2_score(Y_test, y_test_predict))
````
